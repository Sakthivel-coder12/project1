## Data cleanting

```python
df.isnull().sum()   ## finding the null values columns vise 
```

```python
df[df.isnull().any(axis=1)]    ## finding the null values row vise
```

```python
df.loc[:122,"Region"] = 0
df.loc[122:,"Region"] = 1
```

```python
df['Region']= df['Region'].astype(int)
```

```python
df.tail()
```

```python
df = df.dropna().reset_index(drop = True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
## fix spaces int the column name
df.columns.str.replace(" ","")
```

## need to convert the object in to an float or int

```python
## need to convert the object in to an float or int
df.head()
```

```python
df[['month','day','Temperature','RH','Ws']] = df[['month','day','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
df[['Rain', 'FFMC','DMC', 'DC','ISI','BUI','FWI']].astype(float)  ## we can also do it by for loop
```

```python
objects = [features for features in df.columns if df[features].dtypes == 'O']
objects
```

```python
for i in objects:
    if i!="Classes":
        df[i]=df[i].astype(float)
```

## The time more fire will happen(like summer season)

```python
df['Classes'] = np.where(df['Classes'].str.strip() == 'not fire','not fire','fire')
```

```python
## Monthly analysis
dftemp = df.loc[df['Region']==1]
plt.figure(figsize=(20,10))
sns.set_style('whitegrid')
sns.countplot(x = 'month',hue='Classes',data=dftemp)
plt.title("Fire Analysis for sidi - Bel",weight = 'bold')
plt.xlabel("Months",weight = 'bold')
plt.ylabel("Number of Fires",weight = 'bold')
```

```python
## Monthly analysis
dftemp = df.loc[df['Region']==0]
plt.figure(figsize=(20,10))
sns.set_style('whitegrid')
sns.countplot(x = 'month',hue='Classes',data=dftemp)
plt.title("Fire Analysis for Brjaia Regions",weight = 'bold')
plt.xlabel("Months",weight = 'bold')
plt.ylabel("Number of Fires",weight = 'bold')
```
